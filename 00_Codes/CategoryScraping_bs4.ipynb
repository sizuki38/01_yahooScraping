```python
from bs4 import BeautifulSoup
from datetime import datetime
import requests, pandas as pd, os, pickle, time, schedule
```

```python
url     = 'https://news.yahoo.co.jp/ranking/access/news/'
comment = '/comments'
page    = '?page='
order   = '&order=older'
headers = {'User-Agent':'Mozilla/5.0'}
```

```python
def get_list():
    now = datetime.now()
    path = f'./data/{now.strftime("%m-%d")}'
    category = ['','domestic','world','business','entertainment','sports',
                 'it-science','life','local'] 
    for c in category:
        d_ = {}
        response = requests.get(url+c, headers = headers)
        soup = BeautifulSoup(response.content,"html.parser")
        try:
            c_ = soup.title.text.split()[2].split("）")[0]
            os.makedirs(f'{path}/{c_}',exist_ok=True)
            print(f'webtitle: {c_}')
        except:
            pass
        response.close()
        titles = soup.find_all('div', class_ = 'newsFeed_item_title')
        medias  = soup.find_all('span', class_ = 'newsFeed_item_media')
        links  = soup.find_all('a',class_ = 'newsFeed_item_link')
        for title, media, link in zip(titles, medias, links):
            print(f'title: {title.text}')
            print(f'media: {media.text}')
            print(f'link : {link.get("href")}')
            d_[title.text.replace('\u3000', ' ')] = [link.get('href'),media.text]
            if c == '':
                c = 'total'
            with open(f'{path}/{c_}/list_{c}.pkl',"wb") as f:
                pickle.dump(d_,f)
```

```python
def get_comment():
    category = {'総合':'total','国内':'domestic','国際':'world',
                '経済':'business','エンタメ':'entertainment','スポーツ':'sports',
                 'IT・科学':'it-science','ライフ':'life','地域':'local'}
    for c in category:
        print(c)
        with open(f'./data/{now.strftime("%m-%d")}/{c}/list_{category[c]}.pkl', "rb") as f:
            d_ = pickle.load(f)
        for title in d_:
            print(d_[title][1])
            print(f'{d_[title][0]}{comment}')
            response = requests.get(f'{d_[title][0]}{comment}', headers = headers)
            soup = BeautifulSoup(response.content,"html.parser")
            response.close()
            iframe = soup.find('iframe')
            print(iframe)
#             iframe_response = requests.get(src)
#             iframe_soup = BeautifulSoup(iframe_response.content,"html.parser")
```
