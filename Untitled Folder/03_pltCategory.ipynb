```python
import glob, pickle, os, pandas as pd
import requests, matplotlib.pyplot as plt, japanize_matplotlib
from collections import Counter
result_c = 'results/category'
category = {'総合':'total','国内':'domestic','国際':'world',
                '経済':'business','エンタメ':'entertainment','スポーツ':'sports',
                 'IT・科学':'it-science','ライフ':'life','地域':'local'}
```

```python
for c in category:
    data_all = []
    paths = glob.glob(f'data/*/{c}/NLA_data/*')
    print(f'{c}:{len(paths)}')
    for path in paths:
        try:
            with open(path, "rb") as f:
                data = pickle.load(f)
        except Exception as e:
            print(e)
        try:
            data_all += list(data.values())
        except:
            pass
    print(len(data_all))
    scores = Counter(data_all)
    print(sorted(scores.items()))
    plt.bar(scores.keys(),scores.values(),width=0.08)
    plt.title(f'{c}')
    plt.show()
    plt.savefig(f'{result_c}/{c}.png')
```

```python
df = pd.DataFrame()
for c in category:
    data_all = []
    paths = glob.glob(f'data/*/{c}/NLA_data/*')
    for path in paths:
        try:
            with open(path, "rb") as f:
                data = pickle.load(f)
        except:
            pass
        try:
            data_all += list(data.values())
        except:
            pass
    
    df_c = pd.DataFrame(
        Counter(data_all).items(),
        columns=['score',c]
    )
    print(df_c)
    try:
        df = pd.merge(df,df_c)
    except Exception as e:
        print(e)
        df = df_c
df = df.set_index('score').sort_index()
print(df)
```

```python
df_n = pd.DataFrame(columns=category, index=['negative','0','positive'])
for c in category:
    df_n[c]['negative'] = df[c][-0.9:-0.1].sum()
    df_n[c]['positive'] = df[c][0.1:0.9].sum()
    df_n[c]['0'] = df[c][0].sum()
    
df_n 
```

```python
#割合化
for c in category:
    df_n[c] = df_n[c]/df_n[c].sum()*100
    
#グラフオブジェクトの定義し、グラフのサイズを設定
fig, ax = plt.subplots(figsize=(10, 6))

#件名別の積み上げ棒グラフの作成
for i in range(len(df_n)):
    ax.bar(df_n.columns,
          df_n.iloc[i],
          bottom=df_n.iloc[:i].sum(),
          width=.5,
          alpha=.7)

#グラフのタイトル・列ラベルを設定
# ax.set_title('')
ax.set_xlabel('カテゴリ')
ax.set_ylabel('割合(%)')

#グラフの凡例を設定
# ax.legend(df.index.tolist(), loc='lower right', borderaxespad=2)
ax.legend(df_n.index.tolist(), bbox_to_anchor=(1.01, 0.85))
```
