```python
from glob import glob
from collections import Counter
import pickle, matplotlib.pyplot as plt
```

```python
# category = {'総合':'total','国内':'domestic','国際':'world',
#                 '経済':'business','エンタメ':'entertainment','スポーツ':'sports',
#                  'IT・科学':'it-science','ライフ':'life','地域':'local'}
category = {'IT・科学':'it-science','ライフ':'life'}
for i in category:
    datas = []
    paths = glob(f'data/*/{i}/NLA_data/*')
    for path in paths:
        with open(path, "rb") as f:
            data = pickle.load(f)
        datas += data
    scores =[i[1] for i in datas]
    print(len(scores))
    score_all=Counter(scores)
    print(score_all)
    score_k = [k for k,v in score_all.items()]
    score_v = [v for k,v in score_all.items()]
    print(i)
    plt.bar(score_k,score_v)
    plt.show()
```
