```python
for c in category:
    NLA = {}
    paths = glob(f'data/*/{c}/comments/*')
    titles = [os.path.splitext(os.path.basename(p))[0] for p in paths]
    for title in titles:
        title += '。'
        Retitle = title.replace('。。','。')
        
        body = {
            "document": {
                "type": "PLAIN_TEXT",
                "language": "JA",
                "content": Retitle},
            "encodingType": "UTF8"
        }
        
        try:
            res = requests.post(url, headers=header, json=body).json()
        except:
            time.sleep(10)
            try:
                res = requests.post(url, headers=header, json=body).json()
            except:
                pass
            else:
                print(f'{res["documentSentiment"]["score"]} : {title}')
                NLA[text] =  res["documentSentiment"]["score"]
#                 time.sleep(1)
        else:
            print(f'{res["documentSentiment"]["score"]} : {title}')
            NLA[title] =  res["documentSentiment"]["score"]
#             time.sleep(1)
    with open('results/title/NLA_data_{c}',"wb")as f:
        pickle.dump(NLA,f)
```
