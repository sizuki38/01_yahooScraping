```python
from datetime import datetime
import time, schedule, re

from glob import glob
from datetime import datetime
import os, pickle

from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By

options = webdriver.ChromeOptions()
options.add_argument('--headless')
driver = webdriver.Chrome(ChromeDriverManager().install(), options=options)

url     = 'https://news.yahoo.co.jp/ranking/access/news/'
comment = '/comments'
page    = '?page='
order   = '&order=newer'
```

```python
get_allcomment()
```

```python
def task1():
    get_comment()
```

```python
schedule.every().day.at("23:30").do(task1)
```

```python
while True:
    schedule.run_pending()
    time.sleep(1)
```

```python
def get_allcomment():
    now = datetime.now()
    paths = glob('data/*')
#     paths = ['data/06-17','data/06-18','06-19']
    category1 = {'総合':'total','国内':'domestic','国際':'world'}
    category2 = {'経済':'business','エンタメ':'entertainment','スポーツ':'sports'}
    category3 = {'IT・科学':'it-science','ライフ':'life','地域':'local'}
    
    for path in paths:
        print(path)
        for c in category3:
            os.makedirs(f'{path}/{c}/comments',exist_ok=True)
    #         os.makedirs(f'{path}/{c}/news',exist_ok=True)
            with open(f'{path}/{c}/list_{category3[c]}.pkl', "rb") as f:
                d_ = pickle.load(f)
            for title in d_:
                print(title)
                driver.get(f'{d_[title][0]}{comment}')
                comments = []
                try:
                    iframe = driver.find_element(By.CLASS_NAME,'news-comment-plguin-iframe')
                    driver.switch_to.frame(iframe)
                except:
    #                 print(title)
    #                 print(f'{d_[title][0]}{comment}')
                    print('this news is none comment.')
                else:
                    try:
                        count = driver.find_element(By.CLASS_NAME,'counter').text.split('/')[-1].split('件')[0]
                    except:
                        print('pass1')
                    else:
                        for i in range(int(int(count)/10)+1)[::-1]:
#                         print(f'     {d_[title][0]}{comment}{page}{i+1}{order}')
                            driver.get(f'{d_[title][0]}{comment}{page}{i+1}{order}')
                            try:
                                iframe = driver.find_element(By.CLASS_NAME,'news-comment-plguin-iframe')
                                driver.switch_to.frame(iframe)
                                cmtBodys = driver.find_elements(By.CLASS_NAME,'cmtBody')
                                comments += [cmtBody.text for cmtBody in cmtBodys]
                            except:
                                print('pass2')
                            print(comments)
                        with open(f'{path}/{c}/comments/{title.replace("/", "-")}.pkl',"wb")as f:
                            pickle.dump(comments,f)
                    
```
